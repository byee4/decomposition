````python
%matplotlib notebook

import bokeh

from decomposition import PCAPlotter
from decomposition import ExpressionTable
from decomposition import Experiment
from decomposition import color_helpers as ch


import pandas as pd
import numpy as np

# bokeh stuff
from bokeh.io import push_notebook, show, output_notebook # for plotting inline
from bokeh.plotting import figure, output_file, Figure

# from bokeh.plotting import Figure, show, output_file # for saving to file
from ipywidgets import interact
from bokeh.models import HoverTool

output_notebook()
````

````python
counts = 'data/scrna_matrix.txt'
genes_of_interest = 'data/scrna_test_markers.txt'
starting_gene = 'ENSG00000118271'

expt = Experiment.Experiment(
    counts_file=counts,
    conditions_file=None,
    conditions_col=None,
    gene_id=starting_gene,
    is_featurecounts=False
)

# [Optional] transform raw count values into log2
expt.counts.as_log2(pseudocount=1)
# [Optional] subset to only perform PCA on a few genes of interest
expt.counts.subset(subset_file=genes_of_interest)

""" Define the tools needed """
hover = HoverTool(
        tooltips=[
            ("index", "@idx")
        ]
    )

def update_expression(f):
    """
    when called, sets a new colormap in r
    """
    r.set_color(f)
    push_notebook()

px = figure(tools=[hover])
r = PCAPlotter.pcaplot(expt, bokeh=True, ax=px, cmap='Purples')
show(px, notebook_handle=True)
````

### Interactively change colors based on gene expression within each cell idx

````python
df = pd.read_table(counts,index_col=0)
markers = sorted(
    list(
        set(
            [line.strip() for line in open(genes_of_interest,'r')]
        ).intersection(
            set(df.index)
        )
    )
)
interact(update_expression, f=markers)
````
